```python
from sklearn.datasets import fetch_20newsgroups

docs = fetch_20newsgroups(
    subset='train',
    remove=('headers', 'footers', 'quotes')
).data[:3000]
```

![STEP 1 - Preprocessing](https://img.shields.io/badge/STEP%201%20-%20Preprocessing-blue)

```python
#comlete the code
import inflect
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
nltk.download('punkt')
nltk.download('punkt_tab')  
nltk.download('stopwords')
p = inflect.engine()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    
    
    text = text.lower()
    # remove white space and non-alphabetic characters
    text = re.sub(r'[^a-z\s]', ' ', text)
    temp = text.split()
    s2=[]
    
    for word in temp:
        if word.isdigit():
            s2.append(p.number_to_words(word))
        else:
            s2.append(word)
    res = ' '.join(s2)
    trans = str.maketrans('', '', string.punctuation)
    res = res.translate(trans)
    res=" ".join(res.split())
    tokens = word_tokenize(res)
    

    
    
    
    
    #tokens = text.split()
    # to remove noise 
    tokens = [t for t in tokens if len(t) > 2]
    tokens = [t for t in tokens if t not in stop_words]
    return tokens
preprocess(docs[10])
```

![STEP 2 - Vocabulary](https://img.shields.io/badge/STEP%202%20-%20Vocabulary-blue)

```python
#comlete the code
def build_vocab(docs):
    
    vocab_set = set()
    for d in docs:
        tokens = preprocess(d)
        vocab_set.update(tokens)
    
    return sorted(vocab_set) #the list of unique words
```

[![STEP 3 DF](https://img.shields.io/badge/STEP_3_DF-Document_Frequency-pink)](https://digitalpro.dev)

```python
from collections import Counter
def compute_df(docs, vocab):
    
    # Build a word -> index map to faster lookup
    #
    word_to_index = {w: i for i, w in enumerate(vocab)}
    df = [0] * len(vocab)

    for doc in docs:
        tokens = set(preprocess(doc))   # unique words per document
        for token in tokens:
            if token in word_to_index:
                df[word_to_index[token]] += 1

    return df
df_results = compute_df(docs, build_vocab(docs))
print(df_results)
```

[![STEP 4 IDF](https://img.shields.io/badge/STEP_4_IDF-Inverse_Document_Frequency-pink)](https://digitalpro.dev)

```python
import math

def compute_idf(df, N):
    
    idf = []
    for df_val in df:
        idf_val = math.log((1 + N) / (1 + df_val)) + 1
        idf.append(idf_val)
    return idf
```

[![STEP 5 TF Vector](https://img.shields.io/badge/STEP_5_TF-Term_Frequency_Vector-pink)](https://digitalpro.dev)

```python
from collections import Counter

def compute_tf(doc, vocab):
    tokens = preprocess(doc)
    total = len(tokens)

    # if empty doc → return all zeros
    if total == 0:
        return [0.0] * len(vocab)

    counts = Counter(tokens)

    tf_vector = []
    for word in vocab:
        tf_vector.append(counts[word] / total)

    return tf_vector
print(compute_tf(docs[100], build_vocab(docs)))
```

[![STEP 6 TF-IDF Matrix](https://img.shields.io/badge/STEP_6_TFIDF-Build_TF_IDF_Matrix-pink)](https://digitalpro.dev)

```python
import numpy as np

def build_tfidf(docs):
   
    vocab = build_vocab(docs)
    N = len(docs)

    df  = compute_df(docs, vocab)
    idf = compute_idf(df, N)
    idf_array = np.array(idf)   # shape: (vocab_size,)

    tfidf_matrix = np.zeros((N, len(vocab)))

    for i, doc in enumerate(docs):
        tf = compute_tf(doc, vocab)
        tfidf_matrix[i] = np.array(tf) * idf_array

    return tfidf_matrix, vocab
```

[![STEP 7 Cosine Similarity](https://img.shields.io/badge/STEP_7-Cosine_Similarity-pink)](https://digitalpro.dev)

```python
def cosine(a, b):
    
    a = np.array(a, dtype=float)
    b = np.array(b, dtype=float)

    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)

    if norm_a == 0.0 or norm_b == 0.0:
        return 0.0

    return float(np.dot(a, b) / (norm_a * norm_b))

def tfidf(doc_elements: dict[str, int], idfs: dict[str, int]) -> dict:
    """TF * IDF per word given a single word in a single document"""

    tfidf_dict = {}

    for word, val in doc_elements.items():
        tfidf_dict[word] = val * idfs[word]

    return tfidf_dict
print(tfidf({'word1': 0.5, 'word2': 0.3}, {'word1': 2.0, 'word2': 3.0}))
```

[![STEP 8 Search Engine](https://img.shields.io/badge/STEP_8-Search_Engine_REAL_VERSION-orange)](https://digitalpro.dev)

```python
import numpy as np
def search(query, docs, top_k=5):
    tfidf_matrix, vocab = build_tfidf(docs)

    q_vec = np.zeros(len(vocab)) # build query vector
    q_words = preprocess(query)

    for i, w in enumerate(vocab):
        q_vec[i] = q_words.count(w)

    if np.sum(q_vec) > 0:
        q_vec = q_vec / np.sum(q_vec)

    scores = []

    for i, doc_vec in enumerate(tfidf_matrix):
        score = cosine(q_vec, doc_vec)
        scores.append((score, i))

    scores.sort(reverse=True)

    return scores[:top_k]
```
